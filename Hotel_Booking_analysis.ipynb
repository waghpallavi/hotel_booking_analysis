```python
# We will write a function that plots a pie chart using grouped count values so that it can be reused in the future.

# Function for Column Value Counts

def column_value_count(df, column_label):
  df_grpd = df[column_label].value_counts()
  df_grpd = pd.DataFrame({'index':df_grpd.index, 'count':df_grpd.values})
  return df_grpd


# function for visulization of pie chart
def plot_pie_chart_from_column(df, column_label):
  df_grpd = column_value_count(df, column_label)
  fig, ax = plt.subplots(figsize=(14, 6))
  ax.pie(df_grpd.loc[:, 'count'], labels=df_grpd.loc[:, 'index'],explode=[0,0.1],autopct='%1.2f%%', startangle=90, colors=['#A2C3DB','#8871A0'])
  plt.show()
```

```python
# plot a pie chart to show % of bookings from each hotel
plot_pie_chart_from_column(ready_df, 'hotel')
```

City Hotel has 1.6 times the number of bookings as Resort Hotel. City Hotel appears to be more popular among visitors.

```python
# Selecting and counting number of cancelled bookings for each hotel.
cancelled_data = ready_df[ready_df['is_canceled'] == 1]
cancel_grp = cancelled_data.groupby('hotel')
D1 = pd.DataFrame(cancel_grp.size()).rename(columns = {0:'total_cancelled_bookings'})

# Counting total number of bookings for each type of hotel
grouped_by_hotel = ready_df.groupby('hotel')
total_booking = grouped_by_hotel.size()
D2 = pd.DataFrame(total_booking).rename(columns = {0: 'total_bookings'})
D3 = pd.concat([D1,D2], axis = 1)
```
